```python
from pycaret.regression import *
#import*ですべてを読み込む
#get_dataで様々なデータを取ってくる
from pycaret.datasets import get_data
```

```python
dataset=get_data('diamond')
```

```python
exp=setup(dataset,target='Price',)
#Priceは目標値,セットアップしてくれている
```

```python
compare_models()
#アルゴリズムの比較をしてくれている　一番上に出てくるのが決定係数順に出てくれて、最適なアルゴリズムを出してくれる
```

```python
model=create_model('et')
#平均とか出してくれる
```

```python
print(model)
```

```python
#ハイパーパラメータチューニング
tuned_model=tune_model(model)
```

```python
predict_model(tuned_model)
#予測
```

```python
plot_model(tuned_model,plot='feature')
#featurで重要度がわかる
```
